```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Data_train_reduced.csv')
```

```python
pd.set_option('display.max_columns',64)
pd.set_option('display.max_rows', 64)
```

```python
data.head()
```

```python
data.drop('Product', axis=1, inplace=True)
missing_data_name = data.isnull()
missing_data_values = data.isnull().sum()
missing_data_dict = [name for name, value in zip(missing_data_name, missing_data_values) if value > 0]
missing_data_dict
```

```python
for item in missing_data_dict:
    data.drop(item, axis=1, inplace=True)
data.isnull().sum()
```

```python
y = data['Instant.Liking']
```

```python
x = data.drop('Instant.Liking', axis=1)
```

```python
data.dtypes
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=10)
model = LogisticRegression(solver='liblinear')
result = cross_val_score(model,x,y,cv=kfold)
```

```python
result.mean()
```

```python
from sklearn.model_selection import GridSearchCV

values_c = np.array([0.01,0.1,1,0.5,1,2,3,10,20,30,50,100])
regularization=['l1', 'l2']
values_grid = {'C':values_c, "penalty":regularization}

grid_regress = GridSearchCV(estimator=model, param_grid=values_c, cv=5)
grid_regress.fit(x,y)
```
